# Premium Game
Aahad and Harshit are playing a premium game. In this game, they are given a set of ‘n’ numbers from 1 to n. In each move, a player will remove any prime number with all its multiples from the set. The player who cannot make a move will lose the game. Aahad will start the game always. Both of them are playing are optimally. You have to tell the winner’s name.
## Input Format:
First line of input will have an integer ‘t’ denoting the number of test cases.
Next ‘t’ lines will have an integer ‘n’ each.
## Output Format:
Each line of test case will have the name of winner either “Aahad” or “Harshit”

### Sample Input:
3<br>
1<br>
2<br>
5<br>
### Sample Output:
Harshit<br>
Aahad<br>
Aahad<br>
